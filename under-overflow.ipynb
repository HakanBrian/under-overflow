```python
import math

def sigma(t):
    return math.exp(t) / (1 + math.exp(t))

def sigma_inv(t):
    return 1 / (1 + math.exp(-t))
```

```python
t_values = [0, 10, -10, 100, -100, 500, -500, 1000, -1000]

print(f"{'t':>8} | {'sigma':>22} | {'sigma_inv':>22}")
print("-"*58)

for t in t_values:
    try:
        s1 = sigma(t)
    except OverflowError:
        s1 = "overflow"

    try:
        s2 = sigma_inv(t)
    except OverflowError:
        s2 = "overflow"

    print(f"{t:8} | {str(s1):>22} | {str(s2):>22}")
```
